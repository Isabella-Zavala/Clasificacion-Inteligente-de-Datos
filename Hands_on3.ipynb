# Hands-on 3  Kmeans

El algoritmo K-means en sklearn es una implementación en la librería de Python Scikit-learn para realizar el agrupamiento (clustering) no supervisado. Su objetivo es dividir un conjunto de datos en un número predeterminado de grupos o clústeres, agrupando puntos de datos similares y minimizando la distancia entre cada punto y el centroide de su clúster.


**1.1  Fundamentos de la técnica**

K-Means es un algoritmo de aprendizaje no supervisado que agrupa datos en k clústeres según su similitud. Cada grupo se forma alrededor de un centroide, y los puntos se asignan al clúster cuyo centroide esté más cercano.
- Objetivo: minimizar la suma de distancias entre los puntos y sus centroides.
- Aplicaciones: segmentación de clientes, compresión de imágenes, análisis exploratorio.


**1.2 Modelo Matemático del KNN**

El algoritmo busca minimizar la función de costo:
$$ J=\sum _{i=1}^k\sum _{x\in C_i}\| x-\mu _i\| ^2
$$$$$$   Donde:
- k: número de clústeres.
- C_i: conjunto de puntos en el clúster i.
- \mu _i: centroide del clúster i.
- \| x-\mu _i\| : distancia euclidiana entre el punto x y el centroide.

**1.3 Descripción de la librería, clases y funciones**
- sklearn.cluster.KMeans: clase principal para aplicar el algoritmo.
- fit(): ajusta el modelo a los datos.
- predict(): asigna clústeres a nuevos datos.
- pandas: manipulación de datos.
- matplotlib y seaborn: visualización.
- numpy: operaciones numéricas.



```python
####Preprocesamiento

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Cargar dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Estadísticas
print(df.describe())

# Gráfico de dispersión
sns.pairplot(df)
plt.suptitle("Gráfico de dispersión de características", y=1.02)
plt.show()
```

```python
#####Prediction

from sklearn.cluster import KMeans

# Crear modelo KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)

# Función de predicción
def predecir_cluster(patron):
    cluster = kmeans.predict([patron])[0]
    print(f"El patrón pertenece al clúster: {cluster}")

# Ejemplo
predecir_cluster([5.1, 3.5, 1.4, 0.2])
```

```python
#####Model evaluation

from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Convertir etiquetas reales
labels_true = iris.target
labels_pred = kmeans.labels_

# Reasignar etiquetas para comparar
le = LabelEncoder()
labels_pred_adjusted = le.fit_transform(labels_pred)

# Matriz de confusión
cm = confusion_matrix(labels_true, labels_pred_adjusted)
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de Confusión")
plt.show()

# Accuracy
acc = accuracy_score(labels_true, labels_pred_adjusted)
print(f"Precisión del modelo (Accuracy): {acc:.2f}")
```
